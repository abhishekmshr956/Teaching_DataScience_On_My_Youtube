```python
!ls ./Datasets/Bollywood_Movies/2010-2019/

movie_df = pd.read_csv('./Datasets/Bollywood_Movies/2010-2019/bollywood_2010-2019.csv')
movie_df_metadata = pd.read_csv('./Datasets/Bollywood_Movies/2010-2019/bollywood_meta_2010-2019.csv')
movie_df_ratings = pd.read_csv('./Datasets/Bollywood_Movies/2010-2019/bollywood_ratings_2010-2019.csv')
movie_df_text = pd.read_csv('./Datasets/Bollywood_Movies/2010-2019/bollywood_text_2010-2019.csv') 
```

```python
Bajpai_filter = movie_df_text['actors'].str.contains('Manoj Bajpayee', na=False)

Bajpai_movie_id = movie_df_text.loc[Bajpai_filter, 'imdb_id'].values

Bajpai_moviename_filter = movie_df['imdb_id'].isin(Bajpai_movie_id)

movie_df.loc[Bajpai_moviename_filter, 'title']
```

```python
AkshayKumar_filter = movie_df_text['actors'].str.contains('Akshay Kumar', na=False)
AkshayKumar_movie_id = movie_df_text.loc[AkshayKumar_filter, 'imdb_id'].values
AkshayKumar_moviename_filter = movie_df['imdb_id'].isin(AkshayKumar_movie_id)
movie_df.loc[AkshayKumar_moviename_filter, 'title']

movie_df.loc[AkshayKumar_moviename_filter]
```
